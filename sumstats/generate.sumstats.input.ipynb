# Bellenguez.NatGenet.2022.Alzheimers_Disease_Dementia.h.tsv

```python
df = pd.read_csv(f'{INDIR}/Bellenguez.NatGenet.2022.Alzheimers_Disease_Dementia.h.tsv', sep = "\t")
select_cols = ["hm_rsid", "hm_other_allele", "hm_effect_allele", "hm_beta", "hm_effect_allele_frequency", "p_value", "n_cas", "n_con"]
```

```python
filtered_df = df[select_cols].rename(columns = {"hm_rsid": "rsid", 
                                                "hm_other_allele": "other_allele", 
                                                "hm_effect_allele": "effect_allele", 
                                                "hm_beta": "beta", 
                                                "hm_effect_allele_frequency": "EAF"})
filtered_df.head()
```

```python
filtered_df.to_csv(f'{INDIR}/Bellenguez.NatGenet.2022.Alzheimers_Disease_Dementia.formatted.tsv', sep = "\t", index = False)
```

# Wightman.NatGenet.2021.Alzheimers.txt

```python
df = pd.read_csv(f'{INDIR}/Wightman.NatGenet.2021.Alzheimers.txt', sep = "\t").rename(columns = {" chr": "chr"})
df.head()
```

```python
# add rsids 
hg19_rsids = pd.read_csv('/anvil/scratch/x-rwang22/1000G/hg19/g1000_eur.bim', sep = "\t", header = None)
hg19_rsids.columns = ["chr", "rsid", "0", "PosGRCh37", "A1", "A2"]
hg19_rsids.head()
```

```python
filtered_df = pd.merge(df, hg19_rsids, on = ["chr", "PosGRCh37"]).rename(columns = {
    "testedAllele" : "effect_allele", 
    "otherAllele" : "other_allele"
})
select_cols = ["rsid", "effect_allele", "other_allele", "z", "p", "N"]
filtered_df = filtered_df[select_cols]
filtered_df.to_csv(f'{INDIR}/Wightman.NatGenet.2021.Alzheimers.formatted.tsv', sep = "\t", index = False)
filtered_df.head()
```
