```python
for key, value in inputs.items():
    # instantite empty df
    ct_snp_count = {}
    # iterate through each cell type 
    INDIR=f'{BASEDIR}/{key}'
    for ct in os.listdir(INDIR):
        CURDIR = f'{INDIR}/{ct}'
        if os.path.isdir(CURDIR):
            # print(ct)
            count = 0
            for chr in glob.glob(f'{CURDIR}/*.annot.gz'):
                df = pd.read_csv(chr, header = 0, compression = 'gzip')
                tmp = df["ANNOT"].sum()
                count = count + tmp
            ct_snp_count[ct] = count
    count_summary = pd.DataFrame.from_dict(ct_snp_count, orient = "index")
    count_summary["90k_Plus"] = count_summary[0] > 90000
    count_summary.to_csv(f'annot_counts/{value}_annot_counts.tsv', sep = "\t")
```
